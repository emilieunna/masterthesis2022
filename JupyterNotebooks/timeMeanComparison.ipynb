```python
sns.set(rc={'figure.figsize':(20,10)})
sns.set_context("talk")
sns.barplot(x="experiment", y="timeTaken", data=df, palette=my_pal)
plt.title("Bar plot of mean execution time for each experiment")
plt.xlabel("Experiment Type")
plt.ylabel("Mean Execution Time (s)")
plt.savefig('/Users/nil/Roskilde Universitet/Sustainable IT - Master thesis Spring 2022 - Documents/General/Experiment/Data/Charts/Comparison/barplotMeanTime.png'
           ,dpi=300, bbox_inches = "tight")
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.set_context("talk")
ax = sns.barplot(x="experiment", y="Mean power", hue="means",
            palette={'grandMeanPowClient': 'white', 'grandMeanPowServer': 'black'}, data=df1)
for bar_group, desaturate_value in zip(ax.containers, [0.5, 1]):
    for bar, color in zip(bar_group, my_pal2):
        bar.set_facecolor(sns.desaturate(color, desaturate_value))
        
hatches = ["///", ""]
for bars, hatch in zip(ax.containers, hatches):
    for bar in bars:
        bar.set_hatch(hatch)
        
plt.title("Bar plot of mean power dissipation for each experiment")
plt.xlabel("Experiment Type")
plt.ylabel("Power Dissipation (W)")

handles, labels = ax.get_legend_handles_labels()

from matplotlib.legend_handler import HandlerTuple
ax.legend(handles=[tuple(bar_group) for bar_group in ax.containers],
          labels=['Client', 'Server'],
          handlelength=4, handler_map={tuple: HandlerTuple(ndivide=None, pad=0.1)})

plt.savefig('/Users/nil/Roskilde Universitet/Sustainable IT - Master thesis Spring 2022 - Documents/General/Experiment/Data/Charts/Comparison/barplotGrandMeanPow_compared.png'
           ,dpi=300, bbox_inches = "tight")
```
